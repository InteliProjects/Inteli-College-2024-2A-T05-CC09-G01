```python
import os
import shutil
```

```python
# Caminho principal do projeto
base_path = r"C:\Users\wasse\Documentos\Projeto-cattle\CattleImageRepository"

# Lista de subpastas que contêm as imagens
subpastas = [
    r"G1\01-08-2024\X", r"G1\01-08-2024\Y",
    r"G1\08-08-2024\X", r"G1\08-08-2024\Y",
    r"G2\06-08\Video_1\X", r"G2\06-08\Video_1\Y",
    r"G2\06-08\Video_2\X", r"G2\06-08\Video_2\Y",
    r"G2\29-07\00000000196000400\X", r"G2\29-07\00000000196000400\Y",
    r"G2\29-07\00000000196000500\X", r"G2\29-07\00000000196000500\Y",
    r"G3\30-07\termica\x_tudo", r"G3\30-07\termica\y_tudo",
    r"G4\02-08\X", r"G4\02-08\Y"
]

# Caminho onde as imagens organizadas serão salvas
output_path = r"C:\Users\wasse\Documentos\Projeto-cattle\OrganizedImages"

# Criar as pastas de saída se não existirem
os.makedirs(os.path.join(output_path, "X"), exist_ok=True)
os.makedirs(os.path.join(output_path, "Y"), exist_ok=True)
```

```python
def process_images(base_path, subpastas, output_path):
    for subpasta in subpastas:
        full_path = os.path.join(base_path, subpasta)
        if "X" in subpasta:
            target_folder = os.path.join(output_path, "X")
        elif "Y" in subpasta:
            target_folder = os.path.join(output_path, "Y")
        
        # Listar todas as imagens na subpasta atual
        images = os.listdir(full_path)
        
        for image in images:
            # Verificar se existe um par correspondente
            corresponding_image = image.replace('_x', '_y') if 'X' in subpasta else image.replace('_y', '_x')
            corresponding_subpasta = subpasta.replace('X', 'Y') if 'X' in subpasta else subpasta.replace('Y', 'X')
            corresponding_full_path = os.path.join(base_path, corresponding_subpasta)
            
            if os.path.exists(os.path.join(corresponding_full_path, corresponding_image)):
                # Mover a imagem para a pasta de destino
                shutil.copy(os.path.join(full_path, image), os.path.join(target_folder, image))
                print(f"Copiando {image} para {target_folder}")

# Processar as imagens
process_images(base_path, subpastas, output_path)
```
