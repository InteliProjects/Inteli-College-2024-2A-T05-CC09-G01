```python
from ultralytics import YOLO
model = YOLO('yolo11n.pt')

# Step 6: Train the model
model.train(
    data='/content/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    workers=4
)

# Passo 7: Validar o modelo
metrics = model.val()

# Opcional: Exibir as mtricas de validao
print(metrics)
```
