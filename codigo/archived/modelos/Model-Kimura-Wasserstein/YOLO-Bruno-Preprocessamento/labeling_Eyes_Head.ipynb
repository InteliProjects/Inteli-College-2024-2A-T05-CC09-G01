```python
import os
import cv2
import numpy as np

# Diretórios
path_to_images_x = "/content/drive/MyDrive/OrganizedImages/Matched/data/images/train"  # Imagens térmicas
path_to_images_y = "/content/drive/MyDrive/OrganizedImages/Matched/Y"  # Máscaras
path_to_labels = "/content/drive/MyDrive/OrganizedImages/Matched/labels_eyes"  # Labels de saída

# Crie o diretório para salvar os labels se ainda não existir
os.makedirs(path_to_labels, exist_ok=True)

# Função para encontrar a bounding box dos olhos (área vermelha na máscara)
def find_eyes(mask_img):
    # Convertemos a imagem para o espaço de cor RGB
    mask_rgb = cv2.cvtColor(mask_img, cv2.COLOR_BGR2RGB)

    # Definimos o intervalo de cor vermelha (valores podem ser ajustados dependendo da cor exata)
    lower_red = np.array([200, 0, 0])
    upper_red = np.array([255, 50, 50])

    # Criar uma máscara para a área vermelha (olhos)
    mask = cv2.inRange(mask_rgb, lower_red, upper_red)

    # Encontrar os contornos da área vermelha
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    eyes_bboxes = []

    for contour in contours:
        # Encontrar a bounding box (caixa delimitadora) ao redor dos olhos
        x, y, w, h = cv2.boundingRect(contour)
        eyes_bboxes.append((x, y, w, h))

    return eyes_bboxes

# Processar todas as imagens térmicas e suas máscaras correspondentes
for img_file in os.listdir(path_to_images_x):
    if img_file.endswith('_x.png'):
        # Carregar imagem térmica e máscara correspondente
        img_x = cv2.imread(os.path.join(path_to_images_x, img_file))
        img_y = cv2.imread(os.path.join(path_to_images_y, img_file.replace('_x.png', '_y.png')))

        # Verificar se as imagens foram carregadas corretamente
        if img_x is None or img_y is None:
            print(f"Erro ao carregar {img_file} ou a máscara correspondente.")
            continue

        # Encontrar as coordenadas dos olhos
        eyes_bboxes = find_eyes(img_y)

        # Preparar o arquivo de label no formato YOLO
        height, width, _ = img_x.shape
        label_file = os.path.join(path_to_labels, img_file.replace('_x.png', '.txt'))

        with open(label_file, 'w') as f:
            for (x, y, w, h) in eyes_bboxes:
                # Converter as coordenadas para o formato YOLO (valores normalizados)
                x_center = (x + w / 2) / width
                y_center = (y + h / 2) / height
                w_norm = w / width
                h_norm = h / height

                # YOLO class 0 será os olhos
                f.write(f"0 {x_center} {y_center} {w_norm} {h_norm}\n")

        print(f"Labels para {img_file} salvos com sucesso.")
```

```python
import os
import cv2
import numpy as np

# Diretórios
path_to_images_x = "/content/drive/MyDrive/OrganizedImages/Matched/data/images/train"  # Imagens térmicas
path_to_images_y = "/content/drive/MyDrive/OrganizedImages/Matched/Y"  # Máscaras
path_to_labels = "/content/drive/MyDrive/OrganizedImages/Matched/labels_head_and_eyes"  # Labels de saída

# Crie o diretório para salvar os labels se ainda não existir
os.makedirs(path_to_labels, exist_ok=True)

# Função para encontrar a bounding box dos olhos (área vermelha na máscara)
def find_eyes(mask_img):
    # Convertemos a imagem para o espaço de cor RGB
    mask_rgb = cv2.cvtColor(mask_img, cv2.COLOR_BGR2RGB)

    # Definimos o intervalo de cor vermelha (valores podem ser ajustados dependendo da cor exata)
    lower_red = np.array([200, 0, 0])
    upper_red = np.array([255, 50, 50])

    # Criar uma máscara para a área vermelha (olhos)
    mask = cv2.inRange(mask_rgb, lower_red, upper_red)

    # Encontrar os contornos da área vermelha
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    eyes_bboxes = []

    for contour in contours:
        # Encontrar a bounding box (caixa delimitadora) ao redor dos olhos
        x, y, w, h = cv2.boundingRect(contour)
        eyes_bboxes.append((x, y, w, h))

    return eyes_bboxes

# Função para encontrar a bounding box da cabeça (área branca na máscara)
def find_head(mask_img):
    # Definimos o intervalo de cor branca para a cabeça
    lower_white = np.array([200, 200, 200])
    upper_white = np.array([255, 255, 255])

    # Criar uma máscara para a área branca (cabeça)
    mask = cv2.inRange(mask_img, lower_white, upper_white)

    # Encontrar os contornos da área branca
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    head_bboxes = []

    for contour in contours:
        # Encontrar a bounding box (caixa delimitadora) ao redor da cabeça
        x, y, w, h = cv2.boundingRect(contour)
        head_bboxes.append((x, y, w, h))

    return head_bboxes

# Processar todas as imagens térmicas e suas máscaras correspondentes
for img_file in os.listdir(path_to_images_x):
    if img_file.endswith('_x.png'):
        # Carregar imagem térmica e máscara correspondente
        img_x = cv2.imread(os.path.join(path_to_images_x, img_file))
        img_y = cv2.imread(os.path.join(path_to_images_y, img_file.replace('_x.png', '_y.png')))

        # Verificar se as imagens foram carregadas corretamente
        if img_x is None or img_y is None:
            print(f"Erro ao carregar {img_file} ou a máscara correspondente.")
            continue

        # Encontrar as coordenadas da cabeça e dos olhos
        head_bboxes = find_head(img_y)
        eyes_bboxes = find_eyes(img_y)

        # Preparar o arquivo de label no formato YOLO
        height, width, _ = img_x.shape
        label_file = os.path.join(path_to_labels, img_file.replace('_x.png', '.txt'))

        with open(label_file, 'w') as f:
            # Salvar as coordenadas da cabeça (classe 0)
            for (x, y, w, h) in head_bboxes:
                x_center = (x + w / 2) / width
                y_center = (y + h / 2) / height
                w_norm = w / width
                h_norm = h / height
                f.write(f"0 {x_center} {y_center} {w_norm} {h_norm}\n")

            # Salvar as coordenadas dos olhos (classe 1)
            for (x, y, w, h) in eyes_bboxes:
                x_center = (x + w / 2) / width
                y_center = (y + h / 2) / height
                w_norm = w / width
                h_norm = h / height
                f.write(f"1 {x_center} {y_center} {w_norm} {h_norm}\n")

        print(f"Labels para {img_file} salvos com sucesso.")
```
